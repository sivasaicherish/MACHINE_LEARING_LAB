```python
# Create a vector (array) of 1XN dimension representing N-
# dimensional feature vector of a sample. Write a program to
# compute the mean and variance of the elements present in the
# array.
```

```python
import pandas as pd

# Sample data
final_df = pd.read_csv("final_dataframe.csv")

# Compute mean and variance for each sample
mean_values = []
variance_values = []

for index, row in final_df.iterrows():
    sample_data = row.values
    n = len(sample_data)  # Number of elements in the sample
    mean = sum(sample_data) / n
    
    # Sample variance calculation
    squared_diff_sum = sum((x - mean) ** 2 for x in sample_data)
    variance = squared_diff_sum / (n - 1)  # Sample variance formula
    
    mean_values.append(mean)
    variance_values.append(variance)

# Add mean and variance columns to the DataFrame
final_df['mean'] = mean_values
final_df['sample_variance'] = variance_values

final_df
```

```python
import pandas as pd

# Sample data
vector1 = final_df['total_bill']
vector2 = final_df['tip']

# Compute mean for each vector
mean_vector1 = vector1.mean()
mean_vector2 = vector2.mean()

# Compute covariance for the sample
covariance = sum((vector1[i] - mean_vector1) * (vector2[i] - mean_vector2) for i in range(len(vector1))) / (len(vector1) - 1)

print("Covariance between the two vectors:", covariance)
```

```python
import pandas as pd
import math



# Compute means for each vector
mean_vector1 = vector1.mean()
mean_vector2 = vector2.mean()

# Compute the sum of the products of the differences
sum_of_products = sum((vector1[i] - mean_vector1) * (vector2[i] - mean_vector2) for i in range(len(vector1)))

# Compute the sum of the squares of the differences for each vector
sum_of_squares_vector1 = sum((x - mean_vector1) ** 2 for x in vector1)
sum_of_squares_vector2 = sum((y - mean_vector2) ** 2 for y in vector2)

# Compute the square root of the sums of squares
sqrt_sum_of_squares_vector1 = math.sqrt(sum_of_squares_vector1)
sqrt_sum_of_squares_vector2 = math.sqrt(sum_of_squares_vector2)

# Compute correlation
correlation = sum_of_products / (sqrt_sum_of_squares_vector1 * sqrt_sum_of_squares_vector2)

print("Correlation between the two vectors:", correlation)
```

# what are Variance Covariance , Correlation

1. Variance:
   - Variance measures the spread or dispersion of a set of data points around their mean. 
   - It is calculated by taking the average of the squared differences between each data point and the mean.
   - Variance provides insight into the variability or volatility of the data.

2. Covariance:
   - Covariance measures the degree to which two variables change together.
   - It indicates the direction of the linear relationship between two variables.
   - A positive covariance indicates that as one variable increases, the other tends to increase as well, while a negative covariance indicates that as one variable increases, the other tends to decrease.

3. Correlation:
   - Correlation is a standardized measure of the linear relationship between two variables.
   - It ranges from -1 to 1, where 1 indicates a perfect positive linear relationship, -1 indicates a perfect negative linear relationship, and 0 indicates no linear relationship.
   - Correlation is unitless and scale-invariant, making it easier to interpret than covariance.
   - Correlation is calculated by dividing the covariance by the product of the standard deviations of the two variables.

In summary, while covariance measures the direction of the relationship between two variables, correlation additionally measures the strength and direction of that relationship, and variance provides insight into the spread of data points around their mean.

```python
import pandas as pd
import numpy as np

# Sample DataFrame

df = final_df

# Convert DataFrame to matrix
matrix = df.drop(columns=['Unnamed: 0']).values.T

# Compute covariance matrix
covariance_matrix = np.cov(matrix)

# Compute correlation matrix
correlation_matrix = np.corrcoef(matrix)

print("Covariance Matrix:")
print(covariance_matrix)
print("\nCorrelation Matrix:")
print(correlation_matrix)
```


This program:
1. Converts the DataFrame into a matrix where each column represents a feature and each row represents a sample.
2. Computes the covariance matrix using `np.cov()` function.
3. Computes the correlation matrix using `np.corrcoef()` function.

Takeaways from these matrices:
- Covariance Matrix:
  - Positive values indicate a positive relationship between features.
  - Negative values indicate a negative relationship between features.
  - Higher absolute values indicate stronger relationships.
- Correlation Matrix:
  - Values closer to 1 or -1 indicate stronger linear relationships.
  - Values closer to 0 indicate weaker linear relationships or no linear relationship.
  - The diagonal entries will all be 1, representing the perfect correlation of each feature with itself.
