### Q1:Create a figure object called fig using plt.figure() , Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis.

```python
import matplotlib.pyplot as plt

fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
data_x = [1, 2, 3, 4, 5]
data_y = [2, 4, 6, 8, 10]
axes.plot(data_x, data_y)

axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
axes.set_title('Plot on Custom Axis')
plt.show()
```

### Q2: Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.

```python
import matplotlib.pyplot as plt

fig = plt.figure()

axes_main = fig.add_axes([0, 0, 1, 1])

axes_inset = fig.add_axes([0.2, 0.5, 0.2, 0.2])

data_x_main = [1, 2, 3, 4, 5]
data_y_main = [2, 4, 6, 8, 10]
axes_main.plot(data_x_main, data_y_main, label='axes_main')

data_x_inset = [1, 2, 3, 4, 5]
data_y_inset = [10, 8, 6, 4, 2]
axes_inset.plot(data_x_inset, data_y_inset, color='orange', label='axes_inset')

axes_main.set_xlabel('X-axis (axes_main)')
axes_main.set_ylabel('Y-axis (axes_main)')
axes_main.set_title('Plot on axes_main')
axes_main.legend()

axes_inset.set_xlabel('X-axis (axes_inset)')
axes_inset.set_ylabel('Y-axis (axes_inset)')
axes_inset.set_title('Plot on axes_inset')
axes_inset.legend()

plt.show()
```

### Q4: Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data.txt'
column_names = ['X', 'Y']

df = pd.read_csv(file_path, delim_whitespace=True, header=None, names=column_names)

print(df)

plt.plot(df['X'], df['Y'])
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Plot with Labels')

plt.show()
```

### Q5: Write a Python program to plot two or more lines with legends, different widths and colors.

```python
import matplotlib.pyplot as plt

data_x = [1, 2, 3, 4, 5]
data_y1 = [2, 4, 6, 8, 10]
data_y2 = [1, 2, 1, 2, 1]

plt.plot(data_x, data_y1, color='red', linewidth=2, label='Line 1')

plt.plot(data_x, data_y2, color='blue', linewidth=3, label='Line 2')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines with Legends')

plt.legend()

plt.show()
```

### Q6 Write a Python program to plot quantities which have an x and y position.

### Q7 Create plot using subplots(), scatter, and some other methods on the Axes object:

```python
import matplotlib.pyplot as plt
import numpy as np

x_positions = [1, 2, 3, 4, 5]
y_positions = [3, 5, 2, 8, 4]

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].scatter(x_positions, y_positions, color='blue', marker='o')
axes[0].set_title('Scatter Plot')
axes[0].set_xlabel('X-axis')
axes[0].set_ylabel('Y-axis')

x_line = np.linspace(1, 5, 100)
y_line = x_line**2
axes[1].plot(x_line, y_line, color='red', label='y = x^2')
axes[1].set_title('Line Plot')
axes[1].set_xlabel('X-axis')
axes[1].set_ylabel('Y-axis')
axes[1].legend()

plt.tight_layout()
plt.show()
```

### Q8: Generate random data for x and y and create a scatter plot using Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
num_points = 50
data_x_positions = np.random.rand(num_points) * 10
data_y_positions = np.random.rand(num_points) * 20

plt.scatter(data_x_positions, data_y_positions, color='green', marker='o', label='Random Data')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Random Data')

plt.legend()

plt.show()
```

### Q9 Create a bar chart to compare the sales of different products. You can use fictional data or come up with your own dataset.

```python
import matplotlib.pyplot as plt
import numpy as np

products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [350, 500, 200, 700, 400]

plt.bar(products, sales, color='skyblue')

plt.xlabel('Products')
plt.ylabel('Sales (units)')
plt.title('Product Sales Comparison')

plt.show()
```

### Q10 Generate a random dataset and create a histogram using Matplotlib to visualize the distribution of the data.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=20, color='skyblue', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')

plt.show()
```

### Q11 Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create your own.

```python
import matplotlib.pyplot as plt

categories = ['Housing', 'Transportation', 'Food', 'Entertainment', 'Utilities']
expenses = [1200, 400, 600, 300, 200]

plt.pie(expenses, labels=categories, autopct='%1.1f%%', colors=['skyblue', 'lightcoral', 'lightgreen', 'lightpink', 'lightskyblue'], startangle=90)

plt.title('Expense Distribution in Budget')

plt.show()
```
